```python
# TODO 04 - your friend ask for all order data in `California` and `Texas` in 2017 (look at Order Date), send him csv file
df_california_texas_2017 = df[((df['State'] == 'California') | (df['State'] == 'Texas')) & (df['Order Date'].dt.year == 2017)]
df_california_texas_2017.to_csv('california_texas_2017.csv', index=False)
df_california_texas_2017.head()
```

```python
# TODO 05 - how much total sales, average sales, and standard deviation of sales your company make in 2017
df[df['Order Date'].dt.year == 2017]['Sales'].describe()
```

```python
# TODO 06 - which Segment has the highest profit in 2018
df_2018 = df[df['Order Date'].dt.year == 2018]
df_2018.groupby('Segment')['Profit'].sum().sort_values(ascending=False).reset_index()
```

```python
# TODO 07 - which top 5 States have the least total sales between 15 April 2019 - 31 December 2019
start_date = '2019-04-15'
end_date = '2019-12-31'
sales_by_state = df[(df['Order Date'] >= start_date) & (df['Order Date'] <= end_date)].groupby('State').agg({'Sales': 'sum'}).reset_index()
sales_by_state.sort_values(by='Sales').head(5)
```

```python
# TODO 08 - what is the proportion of total sales (%) in West + Central in 2019 e.g. 25% 
df_2019 = df[df['Order Date'].dt.year == 2019].groupby('Region')['Sales'].sum()
total_sales = df_2019.sum()
west_east_sales = df_2019['West'] + df_2019['East']
proportion = west_east_sales / total_sales * 100
print("proportion of total sales (%) in West + Central in 2019 is", proportion)

df_2019.head()
```

```python
# TODO 09 - find top 10 popular products in terms of number of orders vs. total sales during 2019-2020
start_date = '2019-01-01'
end_date = '2020-12-31'
df_2019_2020 = df[(df['Order Date'] >= start_date) & (df['Order Date'] <= end_date)]
```

```python
# number of orders
df_orders = df_2019_2020.groupby(['Product ID', 'Product Name']).agg({'Order ID': 'count'})
df_orders.rename(columns={'Order ID': 'Order Count'}, inplace=True)
df_orders.reset_index(inplace=True)
df_orders[['Product Name', 'Order Count']].sort_values(by='Order Count', ascending=False).head(10)
```

```python
# total sales
df_orders = df_2019_2020.groupby(['Product ID', 'Product Name']).agg({'Sales': 'sum'})
df_orders.reset_index(inplace=True)
df_orders[['Product Name', 'Sales']].sort_values(by='Sales', ascending=False).head(10)
```

```python
# TODO 10 - plot at least 2 plots, any plot you think interesting :)
# Time Series of Sales
import seaborn as sns
import matplotlib.pyplot as plt

df_monthly_sales = df[['Order Date', 'Sales']].resample('M', on='Order Date').sum()

sns.set_theme(style="darkgrid")
plt.figure(figsize=(10, 6))
sns.lineplot(x=df_monthly_sales.index, y='Sales', data=df_monthly_sales)

plt.title('Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')

plt.show()
```

```python
# Sales by Region

df_sales_by_region = df.groupby('Region')['Sales'].sum().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='Region', y='Sales', data=df_sales_by_region)

plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')

plt.show()
```
